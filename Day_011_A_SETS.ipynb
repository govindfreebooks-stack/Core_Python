## Set operations




```python
# SET FUNCTIONS:
s1=set()
s2=set()
for i in range(10):
    s1.add(i)
print("s1=",s1)
for i in range(3,12):
    s2.add(i)
print("s2=",s2)
```

```python
# UNION:
union=s1.union(s2)
union
```

```python
# OR symbol => UNION
union=s1|s2
print("union of s1 and s2 is:",union)
```

```python
# INTERSECTION:
intersecton=s1.intersection(s2)
intersecton
```

```python
# AND symbol  => INTERSECTION
intersecton=s1&s2
print("intersecton of s1 and s2 is:",intersecton)
```

```python
# DIFFERENCE:
diff=s1.difference(s2)
diff
```

```python
# - symbol  => DIFFERENCE
diff=s1-s2
print("difference of s1 and s2 is:",diff)
```

```python
diff2=s2.difference(s1)
diff2
```

```python
s2-s1
```

```python
# SYMMETRIC DIFFERENCE
# (keep those elements only which are not present in both sets)
symm=s1.symmetric_difference(s2)
symm
```

```python
# ^ - symbol  => SYMMETRIC DIFFERENCE
symm=s1^s2
print("symmetric difference of s1 and s2 is:",symm)
```

```python
# PROPER SUBSET,SUBSET AND SUPER SET:
if s1>s2:
    print("s1 is superset")
elif s1<s2:
    print("s1 is proper subset of s2")
elif s1<=s2:
    print("s1 is subset of s2")
else:
    print("s1 and s2 are same sets")
```
