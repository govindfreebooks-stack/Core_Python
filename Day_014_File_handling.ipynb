### OS.Walk & Os.path.join

```python
# D15_T331_OSWalk_get_filename_in_Subdir
# This code recorsively search for files in dir & Subdir
# Then prints the file name with pathname
# This code will only on PyCharm not in Jupyter
import os

path = 'C:\\Tmp'

for root, directories, files in os.walk(path, topdown=True):
	for name in files:
		print(os.path.join(root, name))
	for name in directories:
		print(os.path.join(root, name))
#F:\Actual_Training
```

### Deleting lines matching in the list

Suppose you want to remove header & footer. You can provide them in list

```python
# this is for Elsevier to convert pdf to kindle format and remove header & footer lines.
# list to store file lines
lines = []
# read file
with open(r"C:\Tmp\Delete_lines.txt", 'r') as fp:
    # read an store all lines into list
    lines = fp.readlines()

# Write file
with open(r"C:\Tmp\Delete_lines_01.txt", 'w') as fp:
    # iterate each line
    for number, line in enumerate(lines):
        # delete line 5 and 8. or pass any Nth line you want to remove
        # note list index starts from 0
        if number not in [4, 7]:
            fp.write(line)
# replace file with original name
#os.replace('C:\Tmp\Delete_lines-text to delete_01.txt', 'D:\Tmp\Delete_lines-text to delete.txt')            
```

## Deleting Lines Matching a text (string)

#### Water mark removal

```python
# this is for Elsevier to remove water mark lines.
import os

with open(r"C:\Tmp\Delete_lines-text to delete.txt", "r") as input:
    with open(r"C:\Tmp\Delete_lines-text to delete_01.txt", "w") as output:
        # iterate all lines from file
        for line in input:
            # if text matches then don't write it
            if line.strip("\n") != "text to delete":
                output.write(line)

# replace file with original name
#os.replace('C:\Tmp\Delete_lines-text to delete_01.txt', 'D:\Tmp\Delete_lines-text to delete.txt')
```

```python
# To speed up the file operation
with open(r"C:\Tmp\Delete_lines-text to delete.txt", "r") as fp:
    lines = fp.readlines()

with open(r"C:\Tmp\Delete_lines-text to delete02.txt", "w") as fp:
    for line in lines:
        if line.strip("\n") != "text to delete":
            fp.write(line)
```

```python
# This is for Tamilnadu padanool niruvanam wanted to remove "Nadar" "Thevar" from the text books
import os

original_file = r"C:\Tmp\Delete_Texts.txt"
temp_file = r"c:\Tmp\temp.txt" # Please note that you need to have temp.txt file already

string_to_delete = ['Patil', 'Pawar']
with open(original_file, "r") as input:
    with open(temp_file, "w") as output:
        for line in input:
            for word in string_to_delete:
                line = line.replace(word, "")
            output.write(line)
```

###  open(file, mode='r')

| Mode  | Meaning          | File Must Exist? | Description                                  |
| ----- | ---------------- | ---------------- | -------------------------------------------- |
| `'r'` | **Read**         |  Yes            | Opens file for reading (default)             |
| `'w'` | **Write**        |  No             | Creates a new file or overwrites existing    |
| `'a'` | **Append**       |  No             | Adds content at the end of file              |
| `'x'` | **Create**       |  No             | Creates a new file; errors if file exists    |
| `'b'` | **Binary**       |                 | Used with other modes (e.g., `'rb'`, `'wb'`) |
| `'t'` | **Text**         |                 | Text mode (default; often omitted)           |
| `'+'` | **Read & Write** |  No             | Combines read/write (e.g., `'r+'`, `'w+'`)   |
